## 데이터 전처리

```python
import pandas as pd
df_da = pd.read_csv('data_analysis.persona1_dcinside_agony.csv')
```

```python
def pre(path) : 
    df_da = pd.read_csv(path)
    df_da =df_da.drop(columns='_id')
    df_da['contents'] = df_da['contents'].str.replace(r'\n', ' ')
    df_da['contents'] = df_da['contents'].str.replace('- dc official App', '')
    df_da['contents_all'] = df_da['title']+' '+ df_da['contents']
    return df_da
```

```python
df_da = pre('data_analysis.persona1_dcinside_agony.csv')
df_da.dropna(inplace=True)
```

```python
# tokenizing
from sklearn.feature_extraction.text import TfidfVectorizer

tfidfVectorizer = TfidfVectorizer(ngram_range=(1, 2))
```

```python
X = tfidfVectorizer.fit_transform(df_da['contents_all'])
```

```python
X
```

```python
# 조사 빼기
import re
def regex(x) :
    patterns = '(|다양한 |있다|되고|에 |이 |로 |의 |\.|년띠 |년생 |[0-9]|일 |거야 |인지 |은 |을 |는 |가 |라는 |한테 |랑 |와 |ㅔ)'
    text_regex = re.sub(pattern='adhd자폐증', repl='adhd', string=x)
    text_regex = re.sub(pattern=patterns, repl='', string=x)
    return text_regex

    
```

```python
df_da['contents_all'] = df_da['contents_all'].apply(regex)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
```

```python
cv.fit_transform(df_da['contents_all'])
```

```python
data = cv.get_feature_names_out().tolist()
```

```python
range(len(data)).index
```

```python
df_words = pd.DataFrame(columns=['word'], data=data, index=range(len(data)))
```
